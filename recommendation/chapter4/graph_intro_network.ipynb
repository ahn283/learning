```python
import networkx as nx 
import numpy as np 
import matplotlib.pyplot as plt 

# node list, edge list
node_list = [1, 2, 3, 4, 5]
edge_list = [(1, 1), (1, 2), (1, 3), (1, 4), (1, 5),
             (2, 1), (2, 3), (2, 5),
             (3, 1), (3, 2),
             (4, 1),
             (5, 1), (5, 2)
            ]

# create graph (networkx library)
G = nx.Graph()
G.add_nodes_from(node_list)
G.add_edges_from(edge_list)

# create adjacency matrix : 그래프에서 어느 쪽지점들이 변으로 연결되었는지 나타내는 정사각 행렬
adj_matrix = nx.adjacency_matrix(G).todense()
print("Adjacency Matrix:")
print(adj_matrix)
```

```python
# labels
custom_labels = {1:"Seoul", 2:"Daejon", 3:"Daegu", 4:"Busan", 5:"Gwangju"}

# plot graph
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, labels=custom_labels, node_color='lightblue', node_size=1500, font_size=16, font_weight='bold')
plt.title('Big Cities in Korea')
plt.show()
```

```python
# load Karate club dataset
G = nx.karate_club_graph()

# print the number of nodes and edges
print(f"Nodes: {G.number_of_nodes()}")
print(f"Edges: {G.number_of_edges()}")
```

```python
G.nodes
```

```python
# add a new node with id=100
G.add_node(100)
```

```python
G.nodes
```

```python
G.edges
```

```python
# connect the new node to all existing nodes
for node in G.nodes:
    if node != 100:
        G.add_edge(100, node, weight=1)
```

```python
G.edges
```

```python
# plot circular graph
plt.figure(figsize=(10, 10))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_size=100, font_size=12, font_weight='bold')
plt.title("Karate Club Graph Visualization")
plt.show()
```

```python
# degree of each node
degree_sequence = [d for n, d in G.degree()]
print(f"Degree sequence: {degree_sequence}")
```

```python
[(n, d) for n, d in G.degree()]
```

```python
# average degree
average_degree = np.mean(degree_sequence)
print(f"Average degree: {average_degree}")
```

```python
# min/max degree
max_degree = np.max(degree_sequence)
min_degree = np.min(degree_sequence)
print(f"Max degree: {max_degree}, Min degree: {min_degree}")
```

```python
# distribution of degrees
plt.hist(degree_sequence, bins=range(min(degree_sequence), max(degree_sequence) + 2))
plt.title("Degree Histogram")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
# graph density
density = nx.density(G)
print(f"Density: {density}")
```

```python
# shortest distance between two nodes
path_length = nx.shortest_path_length(G, source=1, target=10)
print(f"Shortest path length from node 1 to node 10: {path_length}")
```

```python
# adjacency matrix
A = nx.adjacency_matrix(G)
print("Adjacency Matrix:")
print(A.todense())
```

```python
# interactive visualization

import networkx as nx 
from community import community_louvain
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# load Les Miserables dataset
G = nx.les_miserables_graph()

# create partitions using Louvain method
partition = community_louvain.best_partition(G)

# set colors to the commnunity
colors = [mcolors.CSS4_COLORS[key] for key in mcolors.CSS4_COLORS.keys()][20:40]
node_colors = [colors[community] for community in partition.values()]

# visualization with spring layout
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=100, font_size=8, font_weight='bold')
plt.title("Les Miserables Visualziation")
plt.show()
```

```python
import plotly.graph_objects as go
import networkx as nx

edge_trace = go.Scatter(
    x=[],
    y=[],
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines'
)

for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_trace['x'] += tuple([x0, x1, None])
    edge_trace['y'] += tuple([y0, y1, None])
    
# create a trace for the nodes
node_trace = go.Scatter(
    x=[],
    y=[],
    text=[],
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='Viridis',
        color=[],
        size=10,
        colorbar=dict(
            thickness=15,
            title='Community',
            xanchor='left',
            titleside='right'
        ),
        line=dict(width=2)
    )
)

for node in G.nodes():
    x, y = pos[node]
    node_trace['x'] += tuple([x])
    node_trace['y'] += tuple([y])
    node_trace['text'] += tuple([f"Chracter: {node} | Community: {partition[node]}"])
    node_trace['marker']['color'] += tuple([partition[node]])

# create a layout for the plot
layout = go.Layout(
    title='Les Miserables Grapth Visualization',
    titlefont=dict(size=16),
    showlegend=False,
    hovermode='closest',
    margin=dict(b=20, l=5, r=5, t=40),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
)

# create a Figure and display the interactive plot
fig = go.Figure(data=[edge_trace, node_trace], layout=layout)
fig.show()
```
