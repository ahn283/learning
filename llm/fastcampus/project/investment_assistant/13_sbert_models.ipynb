```python
from sentence_transformers import SentenceTransformer
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores import Chroma

# 1. Load a pretrained Sentence Transformer model
# we can get the multilingual model from https://www.sbert.net/docs/sentence_transformer/pretrained_models.html#multilingual-models
model_1_name = "distiluse-base-multilingual-cased-v1"
model = SentenceTransformer(model_1_name)

# the sentences to encode
sentences = [
    "The weather is lovely today.",
    "It's so sunny outside!",
    "He drove to the stadium.",
]

# 2. Calculate the embeddings by calling model.encode()
embeddings = model.encode(sentences)

print(embeddings.shape)
# [3, 384]
print(embeddings[0])
```
