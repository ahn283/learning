# Historical stock prices

```python
# import library
import yfinance as yf
import pandas as pd 
from datetime import datetime

# create ticker instance
aapl = yf.Ticker('AAPL')

# get today date and convert it to a string formtted YYYY-MM-DD
end_date = datetime.now().strftime('%Y-%m-%d')
aapl_stock_price = aapl.history(start='2020-01-01', end=end_date)
pd.DataFrame(aapl_stock_price)
```

```python
# getting the max 1 minute data.
aapl_min = appl.history(period='max', end=end_date, interval='1m')
pd.DataFrame(aapl_min)
```

```python
# multiple assets
companies = ['AMZN', 'GOOG', 'WNT', 'TSLA', 'META']
tickers = yf.Tickers(companies)
ticker_prices = tickers.history(period='max', end=end_date, interval='1m')
pd.DataFrame(ticker_prices)
```

```python
# transform multi-level index into a single-index set of columns
ticker_prices.stack(level=1).rename_axis(['Date', 'Ticker']).reset_index(level=1)
```

# Fetch Fundamental Data

```python
import yfinance as yf  

# createa ticker instance of 'Tesla'
tsla = yf.Ticker('TSLA')

# call the functions
tsla_actions = tsla.get_actions()
tsla_analyst = tsla.get_analyst_price_targets()
tsla_balance = tsla.get_balance_sheet()
tsla_calendar = tsla.get_calendar()
tsla_cf = tsla.get_cash_flow()
tsla_info = tsla.get_info()
tsla_inst_holders = tsla.get_institutional_holders()
tsla_news = tsla.get_news()
tsla_recommendation = tsla.get_recommendations()
tsla_sustainability = tsla.get_sustainability()

# print the results
print(tsla_actions)
print('-'*20)
print(tsla_analyst)
print('-'*20)
print(tsla_balance)
print('-'*20)
print(tsla_calendar)
print('-'*20)
print(tsla_cf)
print('-'*20)
print(tsla_info)
print('-'*20)
print(tsla_inst_holders)
print('-'*20)
print(tsla_news)
print('-'*20)
print(tsla_recommendation)
print('-'*20)
print(tsla_sustainability)
print('-'*20)
```

# Fetch Options Chain Data

```python
# import library
import yfinance as yf

# create a ticker of Tesla
tsla = yf.Ticker('TSLA')

# fetch options chain data
tsla_options = tsla.option_chain()

# access both the calls and puts and store then in their respective variables
tsla_puts = tsla_options.puts
tsla_calls = tsla_options.calls
```

## Plotting a Candlestick Chart using yfinance and finplot

```python
# import libraries
import yfinance as yf   # pip install yfinance
import finplot as fplt  # pip install finplot

# create a ticker of Tesla
tsla = yf.Ticker('TSLA')

# retrieve 1 year worth of daily data of Tesla
df = tsla.history(interval='1d', period='1y')

# plot the OHLC candle chart
fplt.candlestick_ochl(df[['Open', 'Close', 'High', 'Low']])
fplt.show()
```
